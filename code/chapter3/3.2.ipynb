# 3.2 时间序列

```python
import csv
import numpy as np
import torch
bikes_numpy = np.loadtxt("../data/chapter3/bike-sharing-dataset/hour-fixed.csv",
                        dtype=np.float32,
                        delimiter=",",
                        skiprows=1,
                        converters={1: lambda x: float(x[8:10])})#将日期字符串转换为与列1中的月份对应的数字
bikes = torch.from_numpy(bikes_numpy)
bikes
```

## 数据重塑

```python
bikes.shape, bikes.stride()
```

```python
daily_bikes = bikes.view(-1, 24, bikes.shape[1])
daily_bikes.shape, daily_bikes.stride()
```

## 数据转置

```python
daily_bikes = daily_bikes.transpose(1, 2)
daily_bikes.shape, daily_bikes.stride()
```

## 独热编码

```python
first_day = bikes[:24].long()
weather_onehot = torch.zeros(first_day.shape[0], 4)
first_day[:,9]
```

```python
weather_onehot.scatter_(
    dim=1,
    index=first_day[:,9].unsqueeze(1) - 1,
    value=1.0)
```

## 数据连接

```python
torch.cat((bikes[:24], weather_onehot), 1)[:1]
```

```python
daily_weather_onehot = torch.zeros(daily_bikes.shape[0], 4, daily_bikes.shape[2])
daily_weather_onehot.shape
```

```python
daily_weather_onehot.scatter_(1, daily_bikes[:,9,:].long().unsqueeze(1) - 1, 1.0)#你将值减少1，因为天气情况的范围是1到4，而索引是从0开始
daily_weather_onehot.shape
```

```python
daily_bikes = torch.cat((daily_bikes, daily_weather_onehot), dim=1)
```

```python
daily_bikes[:, 9, :] = (daily_bikes[:, 9, :] - 1.0) / 3.0
```

## 数据缩放

```python
temp = daily_bikes[:, 10, :]
temp_min = torch.min(temp)
temp_max = torch.max(temp)
daily_bikes[:, 10, :] = (daily_bikes[:, 10, :] - temp_min) / (temp_max -
                                                              temp_min)
```

```python
temp = daily_bikes[:, 10, :]
daily_bikes[:, 10, :] = (daily_bikes[:, 10, :] - torch.mean(temp)) / torch.std(temp)
```
